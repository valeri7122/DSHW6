```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

%matplotlib inline
```

```python
raw_data1 = pd.read_csv('data\data_2d.csv', header=None, names=['1', '2', '3'])
raw_data1.head()
```

```python
scaler = StandardScaler().fit(raw_data1)
norm = scaler.transform(raw_data1)
scaled_data1 = pd.DataFrame(norm)
scaled_data1.head()
```

```python
K = range(1, 10)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_data1)
    meandistortions.append(np.mean(np.min(cdist(scaled_data1, kmeans.cluster_centers_, 'euclidean'), axis=1)))
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Distortion')
plt.title('Метод ліктя data_2d set')
plt.show()
```

```python
raw_data2 = pd.read_csv('data\mnist.csv', header=None)
raw_data2.head()
```

```python
scaler = StandardScaler().fit(raw_data2)
norm = scaler.transform(raw_data2)
scaled_data2 = pd.DataFrame(norm)
scaled_data2.head()
```

```python
K = range(1, 10)
meandistortions = []
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_data2)
    meandistortions.append(np.mean(np.min(cdist(scaled_data2, kmeans.cluster_centers_, 'euclidean'), axis=1)))
plt.plot(K, meandistortions, 'bx-')
plt.xlabel('Кількість кластерів')
plt.ylabel('Distortion')
plt.title('Метод ліктя mnist set')
plt.show()
```

Отже, за домомогою Метода Ліктя можна побачити, що перший набір даних найкраще розділяється на 2 класи, а другий набір данних на 4 класи. Тому для візуалізації ми візьмемо саме такі значення параметру "n_clusters".

```python
kmeans1 = KMeans(n_clusters=2)
kmeans1.fit(scaled_data1)

plt.title('Візуалізація data_2d set')
plt.scatter(scaled_data1.loc[:,1], scaled_data1.loc[:,2],c=kmeans1.labels_)
```

```python
pca = PCA(n_components=4)
scaled_data2_reduce = pca.fit_transform(scaled_data2)

kmeans2 = KMeans(n_clusters=4)
kmeans2.fit(scaled_data2_reduce)

plt.title('Візуалізація mnist set')
plt.scatter(scaled_data2_reduce[:,1], scaled_data2_reduce[:,2], c=kmeans2.labels_)
```
